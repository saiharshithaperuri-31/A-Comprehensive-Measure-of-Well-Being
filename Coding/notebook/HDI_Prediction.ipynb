```python
plt.figure(figsize=(8,5))
sns.histplot(df["Human Development Index (HDI)"], bins=20, kde=True)
plt.title("Distribution of Human Development Index")
plt.xlabel("HDI")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="HUMAN DEVELOPMENT", data=df)
plt.title("Countries by Human Development Category")
plt.xticks(rotation=15)
plt.show()
```

```python
# Remove commas
df["Gross national income (GNI) per capita"] = (
    df["Gross national income (GNI) per capita"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

# Convert to numeric, replacing invalid values like ".." with NaN
df["Gross national income (GNI) per capita"] = pd.to_numeric(
    df["Gross national income (GNI) per capita"],
    errors="coerce"
)

# Fill missing values with the column mean
df["Gross national income (GNI) per capita"].fillna(
    df["Gross national income (GNI) per capita"].mean(),
    inplace=True
)

print("Conversion completed successfully!")
```

```python
plt.figure(figsize=(10,8))

numeric_df = df.select_dtypes(include=["number"])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
